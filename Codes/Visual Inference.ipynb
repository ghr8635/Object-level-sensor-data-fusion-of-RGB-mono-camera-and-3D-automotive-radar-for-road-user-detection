```python
from ultralytics import YOLO
import cv2

# Load a model
model = YOLO(r"D:\AUF-SEM 2\Project\Python\pythonProject1\runs\detect\train33\weights\best.pt")  # pretrained model

# Path to images
images_path = r"D:\AUF-SEM 2\Project\Python\Training Setup\Training Setup\images\predict"

# Run batched inference on a list of images
results = model(images_path)  # return a list of Results objects

# Process results list and display each frame
for result in results:
    frame = result.plot()  
    cv2.imshow('YOLO Predictions', frame)  # Display the frame in a window

    # Wait for a short period of time and check for 'q' key press to exit
    if cv2.waitKey(0) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
