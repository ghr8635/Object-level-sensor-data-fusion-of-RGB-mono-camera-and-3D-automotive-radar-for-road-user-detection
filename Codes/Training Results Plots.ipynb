```python
#Training & Validation Losses

import matplotlib.pyplot as plt
import pandas as pd
# Load the CSV file
rs = pd.read_csv(r"D:\AUF-SEM 2\Project\Python\pythonProject1\runs\detect\train33\results.csv")

# Strip whitespace from column names
#rs.columns = rs.columns.str.strip()

# Calculate the sums for training and validation losses
rs['Train Total Loss'] = rs['         train/box_loss'] + rs['         train/cls_loss']
rs['Validation Total Loss'] = rs['           val/box_loss'] + rs['           val/cls_loss']


plt.figure(figsize=(6, 4))  # Adjust the figure size 

# Plot training and validation total losses
plt.plot(rs['Train Total Loss'], label='Training Loss', color='#e46828')
plt.plot(rs['Validation Total Loss'], label='Validation Loss', color='#1887ab')

# Customize the plot
plt.title('Training & Validation Losses')
plt.xlabel('Epochs')
plt.ylabel('Total Loss')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
# Precision & Recall


import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
rs = pd.read_csv(r"D:\AUF-SEM 2\Project\Python\pythonProject1\runs\detect\train32\results.csv")


plt.figure(figsize=(6, 4))  # Adjust the figure size 

line1, = plt.plot(rs['   metrics/precision(B)'], label='Precision', color='#FFB88A')
line2, = plt.plot(rs['      metrics/recall(B)'], label='Recall', color='#ADD8E6')


# Calculate moving averages
moving_average = 10  #  any average value 
rs['Precision_MA'] = rs['   metrics/precision(B)'].rolling(window=moving_average).mean()
rs['Recall_MA'] = rs['      metrics/recall(B)'].rolling(window=moving_average).mean()

# Plot
line3, = plt.plot(rs['Precision_MA'], label='Precision (Moving Average: 10)', color='#ac1917', linestyle='solid')
line4, = plt.plot(rs['Recall_MA'], label='Recall (Moving Average: 10)', color='blue', linestyle='solid')


# Customize the plot
plt.title('Precision & Recall')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.grid(True)

# Add the legend in the desired order
plt.legend(handles=[line3, line1, line4, line2])

# Show the plot
plt.show()
```
